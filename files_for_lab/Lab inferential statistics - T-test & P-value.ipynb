```python
df2 = pd.read_csv('pokemon.csv',sep=",")
df2
```

```python
p_sample = df2[['Attack','Defense']].sample(50)
p_sample
```

```python
st.ttest_ind(p_sample['Attack'], p_sample['Defense'], equal_var=False)
```

Our hypothesis is that the defense and attack scores are equal. pvalue is greater than 0.05 so null hypothesis stays. which means defense and attack scores are not equal.

```python
df3 = pd.read_excel('anova_lab_data.xlsx')
df3
```

```python
df3['Power_w'] = df3.groupby('Power ').cumcount()
```

```python
df3_pivot = df3.pivot(index='Power_w', columns='Power ', values='Etching Rate')
df3_pivot
```

```python
st.f_oneway(df3_pivot['160 W'],df3_pivot['180 W'],df3_pivot['200 W'])
```

```python
df3_pivot.mean()
```

Our null hypothesis is their means are same. so with that we can understand that there is no effect of power of the plasma beam on etching rate.

Our alternate hypothesis is their means are different. so with that we can understand that there is an effect of power of the plasma beam on etching rate.

the sigficance level is 0.95

In conclusion pvalue is super low than 0.05. so null hypothesis must go.
